```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("shivamb/netflix-shows")

print("Path to dataset files:", path)
```

```python
import pandas as pd
import os

# Load the dataset into a DataFrame 
df = pd.read_csv(os.path.join(path, "netflix_titles.csv"), encoding="latin-1")

# View the data
df.head()
```

```python
# Show info - investigating missing values and data types
df.info()
```

```python
df.dropna(subset=["rating", "duration"], inplace = True)

df.info()
```

```python
df.fillna({"director": "Unknown", "cast": "Unknown", "country": "Unknown"}, inplace = True)

df.info()
```

```python
df.dropna(subset=["date_added"], inplace = True)

df.info()
```

```python
df_no_duplicates = df[~df.duplicated(keep=False)]
df_no_duplicates.info()
```

```python
## No duplicates in the dataset. we are all good

df["date_added"] = pd.to_datetime(df["date_added"].str.strip(), format='%B %d, %Y')

df.info()
```

```python
exempted_cols = ["description", "rating", "duration"]

for col in df.columns:
    if df[col].dtype == "object":
        if col in exempted_cols:
            continue
        df[col] = df[col].str.strip().str.lower()

df.head()
```

```python
df.to_csv("netflix_titles_cleaned.csv", index=False)
```
