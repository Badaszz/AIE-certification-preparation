```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("vivek468/superstore-dataset-final")

print("Path to dataset files:", path)
```

```python
import pandas as pd
import os

# Load the dataset into a DataFrame (I changed the name of the file)
df = pd.read_csv(os.path.join(path, "superstore.csv"), encoding="latin-1")

# View the data
print(df.head())

# Show info
print(df.info())
```

```python
## Filtering

sales_by_region = df.groupby("Region")["Sales"].sum()
print(sales_by_region)
```

```python
profit_high_cap = df["Profit"].quantile(0.95)
high_profit_orders = df[df["Profit"] > profit_high_cap]
print((len(high_profit_orders) / len(df)) * 100)
```

```python
df["Order Date"] = pd.to_datetime(df["Order Date"])
df["Year"] = df["Order Date"].dt.year
sales_by_year = df.groupby("Year")["Sales"].sum()
print(sales_by_year)
```

```python
## Aggregation

sales_by_category = df.groupby("Category")["Sales"].sum()
print(sales_by_category)
```

```python
average_profit_by_region = df.groupby("Region")["Profit"].mean()
print(average_profit_by_region)
```

```python
df["Month"] = df["Order Date"].dt.month
monthly_sales = df.groupby("Month")["Sales"].mean()
Months = {
    1: "January",
    2: "February",
    3: "March",
    4: "April",
    5: "May",
    6: "June",
    7: "July",
    8: "August",
    9: "September",
    10: "October",
    11: "November",
    12: "December"
}

monthly_sales.index = monthly_sales.index.map(Months)
print(monthly_sales)
```

```python
## Pivot tables

pivot_sales_by_region = pd.pivot_table(
    df,
    values = "Sales",
    index = "Region",
    columns = "Category",
    aggfunc = "sum"
)

print(pivot_sales_by_region)
```

```python
profit_by_year = df.groupby("Year")["Profit"].sum()
print(profit_by_year)
```
